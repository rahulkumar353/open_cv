```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Exercise-1:  Perform Bitwise XOR and NOT operation

img1=np.zeros((512,512,3),np.uint8)
img1=cv.rectangle(img1,(200,200),(600,600),(255,255,0),-1)


img2=np.zeros((512,512,3),np.uint8)
(x,y)=(int(img2.shape[0]/2),int(img2.shape[1]/2))
img2=cv.circle(img2,(x,y),(150),(255,255,0),-1)

XOR=cv.bitwise_xor(img1,img2)
NOT=cv.bitwise_not(img1)

plt.imshow(img2)
plt.show()

plt.imshow(img1)
plt.show()

plt.imshow(XOR)
plt.show()

plt.imshow(NOT)
plt.show()


```

```python
# Exercise-2 Using Translation Shift the Image to Down and Right

img=cv.imread('lena.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)

(h, w) = img.shape[:2]          
matrix = np.float32([[1, 0, 100], [0, 1, 200]])

trans = cv.warpAffine(img, matrix, (h,w))

plt.imshow(trans)
plt.show
```

```python
# Exercise-3. Take an Image convert it into HSV and filter the Red Color

img1=cv.imread('lena.jpg')

hsv=cv.cvtColor(img1,cv.COLOR_BGR2HSV)

lower_blue = np.array([50,150,110])
upper_blue= np.array([255,255,255])

mask = cv.inRange(hsv, lower_blue, upper_blue)

res = cv.bitwise_and(img1,img1, mask= mask)

plt.imshow(res)
plt.show
```
